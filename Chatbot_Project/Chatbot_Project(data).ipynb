```python
import pandas as pd
import json
import os

# JSON 파일 로드
with open('BE22000002.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)

# 챗봇 데이터 초기화
chatbot_data = []

# Utterances를 반복하면서 데이터 추출
utterances = data['utterances']
for i in range(0, len(utterances)-1, 2):
    speaker_utterance = utterances[i]['text']
    listener_utterance = utterances[i + 1]['text']

    # 감정 정보 추출 (listener의 감정)
    emotion = utterances[i+1]['listener_empathy']

    # 추출한 정보를 딕셔너리로 구성
    chat_data = {
        'Q': speaker_utterance,
        'A': listener_utterance,
        'Emotion': emotion
    }

    # 챗봇 데이터 리스트에 추가
    chatbot_data.append(chat_data)

# 결과 출력
for i, chat_data in enumerate(chatbot_data, start=1):
    print(f"Dialogue {i}:")
    print(f"Q: {chat_data['Q']}")
    print(f"A: {chat_data['A']}")
    print(f"Emotion: {chat_data['Emotion']}")
    print()

# chatbot_data에 Q, A 및 Emotion이 구성됨
```

```python
# JSON 파일이 저장된 디렉토리 경로
json_directory = '/content'

# 빈 데이터프레임을 생성
combined_df = pd.DataFrame(columns=['Q', 'A', 'Listener_empathy'])

# 디렉토리 내의 모든 JSON 파일을 반복
for filename in os.listdir(json_directory):
    if filename.endswith(".json"):
        file_path = os.path.join(json_directory, filename)

        # JSON 파일 로드
        with open(file_path, 'r', encoding='utf-8') as json_file:
            data = json.load(json_file)

        # 데이터 추출
        chatbot_data = []
        utterances = data['utterances']
        for i in range(0, len(utterances) - 1, 2):
            speaker_utterance = utterances[i]['text']
            listener_utterance = utterances[i + 1]['text']
            listener_empathy = utterances[i + 1]['listener_empathy']

            chat_data = {
                'Q': speaker_utterance,
                'A': listener_utterance,
                'Listener_empathy': listener_empathy
            }
            chatbot_data.append(chat_data)

        # 데이터프레임으로 변환
        df = pd.DataFrame(chatbot_data)

        # 현재 JSON 파일의 데이터프레임을 기존 데이터프레임에 추가
        combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
# 레이블 인코딩
def label_encode_empathy(value):
    if value is None:
        return 0
    elif '조언' in value:
        return 1
    else:
        return 2

# 'listener_empathy' 열에 대해 레이블 인코딩 적용
combined_df['Listener_empathy'].apply(label_encode_empathy)
```

```python
combined_df['Listener_empathy'] = combined_df['Listener_empathy'].apply(label_encode_empathy)
```

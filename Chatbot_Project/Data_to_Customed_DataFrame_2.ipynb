```python
import pandas as pd
import json
import os
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')
```

Relation 종류과 개수 확인

```python
# JSON 파일이 저장된 디렉토리 경로
json_directory = '/content'

# 각 "relation" 종류별 파일 개수를 저장할 딕셔너리 생성
relation_type_counts = defaultdict(int)

# 데이터 디렉토리에서 모든 JSON 파일을 검색
for filename in os.listdir(json_directory):
    if filename.endswith(".json"):
        file_path = os.path.join(json_directory, filename)

        # JSON 파일 열기
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

            # "relation" 값을 가져와서 해당 종류의 개수 증가
            relation = data["info"]["relation"]
            relation_type_counts[relation] += 1

# 각 "relation" 종류별 파일 개수 출력
print("Relation 종류와 각각의 개수 출력: ")
for relation, count in relation_type_counts.items():
    print(f"{relation}: {count} files")
```

```python
# 모든 JSON 파일에서 "relation" 값을 첫 번째 대화 턴(utterance)에 추가

# JSON 파일 로드
for filename in os.listdir(json_directory):
    if filename.endswith('.json'):
        file_path = os.path.join(json_directory, filename)

    # 파일 열기
    with open(file_path, "r") as json_file:
        data = json.load(json_file)

    # 첫 번째 대화 턴(utterance)에 "relation" 추가
    data["utterances"][0]["relation"] = data["info"]["relation"]

    # 수정된 데이터를 다시 파일에 쓰기
    with open(file_path, "w") as json_file:
        json.dump(data, json_file, indent=4)

    print(f"파일 '{filename}'이 업데이트되었습니다.")

print("모든 파일이 업데이트되었습니다.")
```

이제 모든 JSON파일을 하나의 데이터프레임으로 만든다.

```python
# JSON 파일이 저장된 디렉토리 경로
json_directory = '/content'

chatbot_data = []


# JSON 파일 로드
for filename in os.listdir(json_directory):
    if filename.endswith('.json'):
        file_path = os.path.join(json_directory, filename)

        # JSON 파일 로드
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

        # 데이터를 리스트로 만들기
        utterances = data['utterances']

        # 만약 utterance의 길이가 홀수라면, 마지막 홀수 턴은 무시
        if len(utterances) % 2 == 1:
            utterances = utterances[:-1]

        # 데이터 추출
        for i in range(0, len(utterances), 2):

            relation = utterances[0]['relation']
            speaker_utterance = utterances[i]['text']
            speaker_emotion = utterances[i]['speaker_changeEmotion']

            listener_utterance = utterances[i + 1]['text']
            listener_emotion = utterances[i + 1]['listener_empathy']

            if i + 2 >= len(utterances):
                terminate = True
            else:
                terminate = False

            chat_data = {
                'Relation': relation,
                'Q': speaker_utterance,
                'A': listener_utterance,
                'Q_emotion': speaker_emotion,
                'A_emotion': listener_emotion,
                'Terminate': terminate

            }
            chatbot_data.append(chat_data)

# 데이터프레임으로 변환
df = pd.DataFrame(chatbot_data)
```

레이블 인코딩을 위해 Q_emotion과 A_emotion 종류와 개수를 확인한다.

```python
# 'Q_emotion' 열의 각 레이블 개수 확인(None 값 포함)
q_emotion_none_count = df['Q_emotion'].isna().sum()
q_emotion_counts = df['Q_emotion'].value_counts()

print('--------------------------------------------------')
print("Q_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", q_emotion_none_count)
print(q_emotion_counts)


# 'A_emotion' 열의 각 레이블 개수 확인(None 값 포함)
a_emotion_none_count = df['A_emotion'].isna().sum()
a_emotion_counts = df['A_emotion'].value_counts()

print('--------------------------------------------------')
print("A_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", a_emotion_none_count)
print(a_emotion_counts)
```

A_emotion에서
'위로', '동조', '격려', '조언' 총 4개의 레이블로 통합한다.

```python
# 각 행의 'A_emotion' 값을 정리하여 하나의 감정으로 표현

def extract_primary_emotion(emotion_list):
    if isinstance(emotion_list, list) and len(emotion_list) > 0:
        return emotion_list[0]

# 'A_emotion' 열에 적용
df['A_emotion'] = df['A_emotion'].apply(extract_primary_emotion)
```

```python
# 잘 변경되었는지 확인
# 'Q_emotion' 열의 각 레이블 개수 확인(None 값 포함)
q_emotion_none_count = df['Q_emotion'].isna().sum()
q_emotion_counts = df['Q_emotion'].value_counts()

print('--------------------------------------------------')
print("Q_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", q_emotion_none_count)
print(q_emotion_counts)


# 'A_emotion' 열의 각 레이블 개수 확인(None 값 포함)
a_emotion_none_count = df['A_emotion'].isna().sum()
a_emotion_counts = df['A_emotion'].value_counts()

print('--------------------------------------------------')
print("A_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", a_emotion_none_count)
print(a_emotion_counts)
```

````python
import pandas as pd
import json
import os
import warnings
warnings.filterwarnings('ignore')
````

우선, 파일 한 개로 테스트해 본다.

````python
filename = 'BE22000002.json'

# JSON 파일 로드
with open(filename, 'r', encoding='utf-8') as file:
    data = json.load(file)
````

````python
# 데이터를 리스트로 변환
utterances = data["utterances"]
````

````python
# 데이터프레임 초기화
df = pd.DataFrame(columns=["Q_emotion", "Q_utterance", "A_emotion", "A_utterance", "Terminate"])

# 초기값 설정
items = {"Q_emotion": None, "Q_utterance": None, "A_emotion": None, "A_utterance": None, "Terminate": None}
````

````python
for utterance in utterances:
    if utterance["role"] == "speaker":
        items["Q_emotion"] = utterance["speaker_changeEmotion"]
        items["Q_utterance"] = utterance["text"]
    elif utterance["role"] =="listener":
        items["A_emotion"] = ", ".join(utterance["listener_empathy"]) if utterance["listener_empathy"] else None
        items["A_utterance"] = utterance["text"]

    items["Terminate"] = utterance["terminate"]

    # 데이터프레임에 현재 대화 세그먼트 추가
    df = df.append(items, ignore_index=True)
````

이제 모든 JSON파일을 하나의 데이터프레임으로 만든다.

````python
# JSON 파일이 저장된 디렉토리 경로
json_directory = '/content'

# 데이터프레임 초기화
df = pd.DataFrame(columns=["Q_emotion", "Q_utterance", "A_emotion", "A_utterance", "Terminate"])

# 초기값 설정
items = {"Q_emotion": None, "Q_utterance": None, "A_emotion": None, "A_utterance": None, "Terminate": None}


# JSON 파일 로드
for filename in os.listdir(json_directory):
    if filename.endswith('.json'):
        file_path = os.path.join(json_directory, filename)

        # JSON 파일 로드
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

        # 데이터를 리스트로 변환
        utterances = data["utterances"]

        # 초기값 설정
        items = {"Q_emotion": None, "Q_utterance": None, "A_emotion": None, "A_utterance": None, "Terminate": None}

        for utterance in utterances:
            if utterance["role"] == "speaker":
                items["Q_emotion"] = utterance["speaker_changeEmotion"]
                items["Q_utterance"] = utterance["text"]
            elif utterance["role"] =="listener":
                items["A_emotion"] = ", ".join(utterance["listener_empathy"]) if utterance["listener_empathy"] else None
                items["A_utterance"] = utterance["text"]

            items["Terminate"] = utterance["terminate"]

            # 데이터프레임에 현재 대화 세그먼트 추가
            df = df.append(items, ignore_index=True)
````

````python
df
````

````python
drop_copied_df = df.drop_duplicates(subset='Q_utterance')
````

중복된 텍스트가 있는 행을 삭제한다.

레이블 인코딩을 위해 Q_emotion과 A_emotion 종류와 개수를 확인한다.

````python
# 'Q_emotion' 열의 각 레이블 개수 확인(None 값 포함)
q_emotion_none_count = drop_copied_df['Q_emotion'].isna().sum()
q_emotion_counts = drop_copied_df['Q_emotion'].value_counts()

print('--------------------------------------------------')
print("Q_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", q_emotion_none_count)
print(q_emotion_counts)


# 'A_emotion' 열의 각 레이블 개수 확인(None 값 포함)
a_emotion_none_count = drop_copied_df['A_emotion'].isna().sum()
a_emotion_counts = drop_copied_df['A_emotion'].value_counts()

print('--------------------------------------------------')
print("A_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", a_emotion_none_count)
print(a_emotion_counts)
````

````python
# 데이터프레임의 'Q_emotion' 열과 'A_emotion' 열의 종류를 확인한다.
unique_q_emotions = drop_copied_df['Q_emotion'].unique()
unique_a_emotions = drop_copied_df['A_emotion'].unique()

print("Q_emotion 종류:", unique_q_emotions)
print("A_emotion 종류:", unique_a_emotions)
````

````python
# 잘 변경되었는지 확인
# 'Q_emotion' 열의 각 레이블 개수 확인(None 값 포함)
q_emotion_none_count = drop_copied_df['Q_emotion'].isna().sum()
q_emotion_counts = drop_copied_df['Q_emotion'].value_counts()

print('--------------------------------------------------')
print("Q_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", q_emotion_none_count)
print(q_emotion_counts)


# 'A_emotion' 열의 각 레이블 개수 확인(None 값 포함)
a_emotion_none_count = drop_copied_df['A_emotion'].isna().sum()
a_emotion_counts = drop_copied_df['A_emotion'].value_counts()

print('--------------------------------------------------')
print("A_emotion 레이블별 개수:")
print('--------------------------------------------------')
print("None 값 ", a_emotion_none_count)
print(a_emotion_counts)
````

````python
# 레이블 인코딩

q_emotion_mapping = {
    None: 0,
    '중립': 1,
}

a_emotion_mapping = {
    None: 0,
    '조언': 1,
    '위로': 2,
    '격려': 3,
    '동조': 4
}


drop_copied_df['Q_emotion'] = drop_copied_df['Q_emotion'].map(q_emotion_mapping)
drop_copied_df['A_emotion'] = drop_copied_df['A_emotion'].map(a_emotion_mapping)
````

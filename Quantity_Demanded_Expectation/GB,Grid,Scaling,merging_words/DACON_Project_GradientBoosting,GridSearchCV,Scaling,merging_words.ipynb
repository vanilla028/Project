**Dataset Info.**  
**train.csv**    
ID : 학습 샘플의 고유 ID  
수요량 : 학습 샘플의 target  


**test.csv**
ID : 추론 샘플의 고유 ID  

**sample_submission.csv**  - 제출 양식  
ID : 추론 샘플의 고유 ID
수요량 : 수요 샘플의 target


```python
import pandas as pd
import numpy as np
import random
import os
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
words_list = ['과일', '버섯', '홍삼', '한과', '스팸', '곶감', '한우', '새우', '견과', '올리브유', '머스캣', '김', '굴비']

word_dataframes = {}
total_sum = 0

# 반복문을 사용하여 단어별 데이터프레임 생성
for word in words_list:
    filtered_df = train_df[train_df['선물 유형'].str.contains(word, case=False, na=False)]
    word_dataframes[word] = filtered_df
    count = len(filtered_df)
    total_sum += count
    print(f'{word}: {count}개')

print(f'총 합계: {total_sum}개')
```

```python
# 정수 인코딩

# 정수 인덱스를 부여한 딕셔너리 생성
word_to_int = {word: idx for idx, word in enumerate(words_list)}

def map_labels_and_encode(row):
    for word in words_list:
        if word in row['선물 유형']:
            return word_to_int.get(word, -1)  # 해당 단어의 정수 인덱스 반환, 없으면 -1 반환

# '선물 유형' 열을 업데이트하여 정수 인코딩 적용
train_df['선물 유형'] = train_df.apply(map_labels_and_encode, axis=1)

print(train_df['선물 유형'])
```

```python
# 테스트 데이터 정수 인코딩

def map_labels_and_encode(row):
    for word in words_list:
        if word in row['선물 유형']:
            return word_to_int.get(word, -1)  # 해당 단어의 정수 인덱스 반환, 없으면 -1 반환

# '선물 유형' 열을 업데이트하여 정수 인코딩 적용
test_df['선물 유형'] = test_df.apply(map_labels_and_encode, axis=1)

print(test_df['선물 유형'])
```

```python
train_data = train_df.drop(['ID', '수요량'], axis =1)
train_label = train_df['수요량']

test_data = test_df.drop('ID', axis=1)
```

```python
# Label Encoding
nominal_feature = ['쇼핑몰 구분', '도시 유형', '지역 유형', '쇼핑몰 유형']

for feature in nominal_feature:
    le = LabelEncoder()
    le = le.fit(train_data[feature])
    train_data[feature] = le.transform(train_data[feature])

    for label in np.unique(test_data[feature]):
        if label not in le.classes_:
            le.classes_ = np.append(le.classes_, label)
    test_data[feature] = le.transform(test_data[feature])
```

```python
features_to_scale = train_data[['가격(원)', '추석까지 남은 기간(주)']]


# 객체 생성 및 학습
scaler = MinMaxScaler()
scaler.fit(features_to_scale)

normalized_features = scaler.transform(features_to_scale)

train_data[['가격(원)', '추석까지 남은 기간(주)']] = normalized_features
```

```python
# test_data scaling
features_to_scale_test_data = test_data[['가격(원)', '추석까지 남은 기간(주)']]

# 객체 생성 및 학습
scaler_test_data = MinMaxScaler()
scaler_test_data.fit(features_to_scale_test_data)

normalized_features_test_data = scaler.transform(features_to_scale_test_data)

test_data[['가격(원)', '추석까지 남은 기간(주)']] = normalized_features_test_data
```

## 모델링  
GradientBoostingRegressor  
GridSearchCV로 최적의 하이퍼파라미터 찾기

```python
# 그래디언트 부스팅 회귀 모델
model = GradientBoostingRegressor(random_state=42)

# 튜닝할 하이퍼파라미터 그리드 정의
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5]
}

# GridSearchCV를 사용하여 하이퍼파라미터 튜닝
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(train_data, train_label)
```

```python
# 최적의 하이퍼파라미터와 모델 출력
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
print(f'Best Hyperparameters: {best_params}')
```

```python
result_df = pd.read_csv('sample_submission.csv')
result_df['수요량'] = best_model.predict(test_data)
result_df.head()
```
